# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y
```

```python
def pearson_r(x, y):
    """
    Return Pearson correlation coefficient
    """
    corr_mat = np.corrcoef(x, y)
    return corr_mat[0,1]
```

```python
def permutation_sample(data1, data2):
    """Generate a permutation sample from two data sets"""
    
    #concatenate the data sets
    data = np.concatenate((data1, data2))

    #permute the concatenated array
    permuted_data = np.random.permutation(data)

    #split the ermuted array into two
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]
    
    return perm_sample_1, perm_sample_2
```

```python
def draw_perm_reps(data_1, data_2, func, size=1):
    """Generate multiple permutation replicates."""
    
    #Initialize array of replicates
    perm_replicates = np.empty(size)

    for i in range(size):
        #generate permutation sample
        perm_sample_1, perm_sample_2 = permutation_sample(data_1, data_2)

        #compute the test statistic
        perm_replicates[i] = func(perm_sample_1, perm_sample_2)

    return perm_replicates
```

```python
def bootstrap_replicate_1d(data, func):
    """Generate bootstrap replicate of 1d data."""
    bs_sample = np.random.choice(data, len(data))
    return func(bs_sample)
```

```python
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates"""

    #initialize empty array of replicates 
    bs_replicates = np.empty(shape=size)
    
    #Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)
    
    return bs_replicates
```

Hypothesis test 1

Test to see if the correlation between Excess rate of readmission 
and  # of discharges could be by chance. Let's test to see if excess rate of readmission is actually idependent of the # of discharges.

Null Hypothesis: The two variables are uncorrelated

Alternate Hypothesis: The variables are correlated

statistical significance: 0.01

```python
#initialize empty array
perm_reps = np.empty(10000)

for i in range(10000):
    #permute the # of discharges
    discharge_permuted = np.random.permutation(x)
    #compute the correlation
    perm_reps[i] = pearson_r(discharge_permuted, y)
    
p1 = np.sum(perm_reps <= p_r) / len(perm_reps)
print(f'p-value: {p1}')
    
```

```python
_ = plt.hist(perm_reps, density=True, bins=int(np.sqrt(1000)));
_ = plt.axvline(x=p_r, color='red')
_ = plt.xlabel('Pearson-r')
_ = plt.title('PDF')
```

We do not get any values at least as extreme as the actual pearson correlation coefficient, and therefore, reject the null hypothesis that they are uncorrelated.
